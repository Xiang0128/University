```python
import time
import random

# 單人模式
def single_player_mode(rounds=3):
    reaction_times = []
    print("請按 Enter 開始測試...")

    for i in range(rounds):
        input("等待訊號中... (按 Enter 繼續)")
        wait_time = random.uniform(1, 3)
        time.sleep(wait_time)
        print("現在按 Enter!")
        start_time = time.time()
        input()
        end_time = time.time()

        reaction_time = round(end_time - start_time, 3)
        reaction_times.append(reaction_time)
        print(f"第 {i + 1} 次反應時間：{reaction_time} 秒")

    average_time = round(sum(reaction_times) / len(reaction_times), 3)
    print(f"\n平均反應時間：{average_time} 秒")

    return average_time

# 多人模式
def multiplayer_mode():
    players = input("輸入玩家名字（用逗號分隔）：").split(",")
    player_results = {}

    for player in players:
        print(f"\n=== {player.strip()} 的測試 ===")
        average_time = single_player_mode()
        player_results[player.strip()] = average_time

    print("\n============")
    print("最終排名：")

    sorted_results = sorted(player_results.items(), key=lambda x: x[1])
    for rank, (player, avg_time) in enumerate(sorted_results, start=1):
        print(f"{rank}. {player} - {avg_time} 秒")

def main():
    print("歡迎進入反應時間測試遊戲！")
    mode = input("選擇模式：1 單人模式，2 多人模式：")

    if mode == "1":
        print("\n=== 單人模式 ===")
        single_player_mode()
    elif mode == "2":
        print("\n=== 多人模式 ===")
        multiplayer_mode()
    else:
        print("無效的選擇，請重新執行程式。")

if __name__ == "__main__":
    main()
```
