```python
# BMI計算
name = input("輸入姓名：")
height = int(input("輸入身高："))
weight = int(input("輸入體重："))
bmi = weight/((height/100)**2)
print(name+"，你好!\n"+"="*20)
print("你的體重是%6d"%weight,"公斤")
print("你的身高是%6d"%height,"公分\n","="*20)
print("所以你的BMI值是%5.1f"%bmi,"\n","="*20)
print("你的體重",end="")
if (bmi<18.5):
  print("有點過輕喔!!")
elif (bmi>=18.5 and bmi<24):
  print("正常喔!!")
elif (bmi>=24 and bmi<27):
  print("有點過重喔!!")
else:
  print("有點肥胖喔!!")
```

```python
# 計算所得稅額
income = int(input("請輸入所得淨額："))
if(income >= 0 and income <= 540000):
  tax = income*0.05
  print("所需繳納的所得稅額為：%d"%tax)
elif(income >= 540001 and income <= 1210000):
  tax = 540000*0.05 + (income-540000)*0.12
  print("所需繳納的所得稅額為：%d"%tax)
elif(income >= 1210001 and income <= 2420000):
  tax = 540000*0.05 + (1210000-540000)*0.12 + (income-1210000)*0.2
  print("所需繳納的所得稅額為：%d"%tax)
elif(income >= 2420001 and income <= 4530000):
  tax = 540000*0.05 + (1210000-540000)*0.12 + (2420000-1210000)*0.2 + (income-2420000)*0.3
  print("所需繳納的所得稅額為：%d"%tax)
elif(income >= 4530001):
  tax = 540000*0.05 + (1210000-540000)*0.12 + (2420000-1210000)*0.2 + (4530000-2420000)*0.3 + (income-4530000)*0.4
  print("所需繳納的所得稅額為：%d"%tax)
else:
  print("您輸入的所得淨額可能有誤")
```

```python
# 小B、小C、小D成績分別為99、60、50，利用串列記錄成績
grade=[99,60,50]
print(grade)

# 小C成績錯誤，需修改為59
grade[1]=59
print(grade)

# 加入小A與小E的成績，將小A成績記錄在串列的第一筆資料(90分)，小E成績記錄在串列的最後一筆資料(70分)
grade.insert(0,90)
grade.append(70)
print(grade)

# 移除成績最高與最低者
grade.remove(max(grade))
grade.remove(min(grade))
print(grade)

# 改為用元組依序記錄小B、小C、小D成績
t_grade=tuple(grade)
print(t_grade)

# 利用三個串列分別記錄三位學生姓名及其對應的成績，串列中第一個元素為姓名，第二個元素為成績，將三個串列放在一個元組中
a=["小A",90]
b=["小B",59]
c=["小C",50]
t=tuple(a+b+c)
print(t)
```

```python
# 利用迴圈及append()方法，將兩個串列交叉合併
TName = ["阿堂", "阿土", "小張", "小李"]
CName = ["資訊安全", "資料通訊", "資料結構", "資訊管理"]
combined_list = []
for i in range(len(TName)):
    combined_list.append(TName[i])
    combined_list.append(CName[i])
print(combined_list)

# 利用迴圈及insert()方法，將兩個串列交叉合併
TName = ["阿堂", "阿土", "小張", "小李"]
CName = ["資訊安全", "資料通訊", "資料結構", "資訊管理"]
combined_list = []
for i in range(len(TName)):
    combined_list.insert(2*i, CName[i])
    combined_list.insert(2*i+1, TName[i])
print(combined_list)

# 利用extend()方法，將兩個串列連接合併
TName = ["阿堂", "阿土", "小張", "小李"]
CName = ["資訊安全", "資料通訊", "資料結構", "資訊管理"]
combined_list = TName[:]
combined_list.extend(CName)
print(combined_list)

# 利用加法運算，將兩個串列連接合併
TName = ["阿堂", "阿土", "小張", "小李"]
CName = ["資訊安全", "資料通訊", "資料結構", "資訊管理"]
combined_list = CName + TName
print(combined_list)
```

```python
# 使用者可以輸入五筆資料，利用串列將輸入資料依序記錄下來，於輸入完成後:
count=0
l=[]
while(count<5):
  s=input("輸入：")
  l.append(s)
  count+=1

# (1)利用for...in迴圈將串列中的元素內容依序印出來
for i in l:
  print(i,end=" ")

# (2)將串列元素反向(倒轉)排序並印出
l.reverse()
print()
for i in l:
  print(i,end=" ")

# (3)將串列順向(由小到大)排序並印出
l.sort()
print()
for i in l:
  print(i,end=" ")

# (4)將串列逆向(由大到小)排序並印出
l.sort(reverse=True)
print()
for i in l:
  print(i,end=" ")

# (5)利用星號(*)將串列內容連結並印出
print()
print("*".join(l))
```

```python
Sales = ['Python', 400, 'Java', 300, 'C++', 100, 'VB', 200, 'Python', 400, 'C++', 100, 'Python', 400, 'Ruby', 250, 'Delphi', 350]
Book = Sales[::2]
Price = Sales[1::2]
Book_after = []
cost = []
num = []

print("共銷售%d本書" % len(Book))
print("\n")

# 統計每本書的銷售總金額及數量
for i in range(len(Book)):
    if Book[i] not in Book_after:
        Book_after.append(Book[i])
        cost.append(Price[i])
        num.append(1)
    else:
        for j in range(len(Book_after)):
            if Book[i] == Book_after[j]:
                cost[j] += Price[i]
                num[j] += 1

# 輸出每本書的總銷售金額
print("每本書總共銷售金額：")
for i in range(len(Book_after)):
    print(Book_after[i], cost[i])

print("\n")

# 銷售數量最多的書
print("銷售數量最多的書：")
for i in range(len(Book_after)):
    if num[i] == max(num):
        print(Book_after[i], ":", num[i], "本")
        break

print("\n")

# 書名最短的書
min_len = len(Book_after[0])
minnum = 0
print("書名最短的書：")
for i in range(len(Book_after)):
    if len(Book_after[i]) < min_len:
        min_len = len(Book_after[i])
        minnum = i
print(Book_after[minnum], ":", min_len, "字")

# 找出銷售金額最高的書
max_cost = cost[0]
maxnum = 0
for i in range(len(cost)):
    if cost[i] > max_cost:
        max_cost = cost[i]
        maxnum = i

print("\n")
print("銷售金額最高的書：")
print(Book_after[maxnum], ":", max_cost, "元")

# 找出銷售金額最低的書
min_cost = cost[0]
minnum = 0
for i in range(len(cost)):
    if cost[i] < min_cost:
        min_cost = cost[i]
        minnum = i

print("\n")
print("銷售金額最低的書：")
print(Book_after[minnum], ":", min_cost, "元")

# 找出單價最高的書
max_price_per_unit = cost[0] / num[0]
max_price_book = Book_after[0]
for i in range(len(Book_after)):
    price_per_unit = cost[i] / num[i]
    if price_per_unit > max_price_per_unit:
        max_price_per_unit = price_per_unit
        max_price_book = Book_after[i]

print("\n")
print("單價最高的書：")
print(max_price_book, ":", int(max_price_per_unit), "元")

# 找出單價最低的書
min_price_per_unit = cost[0] / num[0]
min_price_book = Book_after[0]
for i in range(len(Book_after)):
    price_per_unit = cost[i] / num[i]
    if price_per_unit < min_price_per_unit:
        min_price_per_unit = price_per_unit
        min_price_book = Book_after[i]

print("\n")
print("單價最低的書：")
print(min_price_book, ":", int(min_price_per_unit), "元")
```
