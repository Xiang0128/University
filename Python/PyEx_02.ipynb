## 範例程式 2-1：計算 BMI

```python
"""
範例程式2-1：計算 BMI 值
"""
# 基本資料輸入區：
name = 'Kevin'                    # 填入姓名，字串
height = 172                     # 填入身高，單位公分，整數
weight = 70                      # 填入體重，單位公斤，整數

# BMI 計算區： BMI = 體重/身高(公尺)平方
bmi = weight / (height/100)**2    # 計算 BMI，浮點數

# 成果輸出區：
print(name, "的 BMI 值是...", bmi)
```

## 範例程式 2-4：`指定`運算子

```python
a = b = c = 10       # 單一值多重指定給多個變數
print(a,b,c)
a, b, c = 1, 2, 3    # 多個值多重指定給對應的多個變數
print(a,b,c)
a, b, c = c, a, b    # 利用多重指定，使多個變數值交換
print(a,b,c)
a, b = b, a          # 利用多重指定，使兩個變數互換值
print(a,b)
```

## 範例程式 2-7：round()

```python
print(round(1.5), round(1.49))               # 小數點前面是奇數
print(round(2.5), round(2.6), round(2.5001)) # 小數點前面是偶數
print(round(0.1234,2), round(5678,-2))       # 進位點不在小數點
```

## 範例程式 2-8：強制轉換型別 `int()`

## 範例程式 2-9：進位數轉換（二進位、八進位、十六進位）

```python
print("246 轉換成二進位是\t",   bin(246),  type(bin(246)) )
print("246 轉換成八進位是\t",   oct(246),  type(oct(246)) )
print("246 轉換成十六進位是\t", hex(246),  type(hex(246)) )
```

## 範例程式 2-12：跳脫字元`\n`

```python
print("你喜歡什麼水果？ A)蘋果 B)芭樂 C)香蕉")
print("-"*40)
print("你喜歡什麼水果？ \n\tA)蘋果 \n\tB)芭樂 \n\tC)香蕉")
print("^-^ "*10)
print('''
你喜歡什麼水果？
    A)蘋果
        B)芭樂
            C)香蕉''')
```

## 範例程式 2-14：輸出 與 格式化符號（簡單版）

```python
name="Peter";  height=175;  weight=70
bmi = weight / (height/100)**2
print("\n%s，你好！"%name)
s1 = "你的體重是 %d公斤"%weight
s2 = "你的身高是 %f公分"
print(s1, s2%height, "—"*30, sep="\n")
print("所以 %s 的BMI值是 %f"%(name, bmi))
```

  <br><br>

## 範例程式 2-15：輸出 與 格式化符號（對齊版）

```python
name="Peter";  height=175;  weight=70
bmi = weight / (height/100)**2
print("\n%s，你好！"%name)
s1 = "你的體重是 %7d公斤"%weight
s2 = "你的身高是 %-7.1f公分"
print(s1, s2%height, "—"*30, sep="\n")
print("所以 %s 的BMI值是 %.2f"%(name, bmi))
```
