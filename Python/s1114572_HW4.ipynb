```python
Sales = ['Python', 400, 'Java', 300, 'C++', 100, 'VB', 200, 'Python', 400, 'C++', 100, 'Python', 400, 'Ruby', 250, 'Delphi', 350]
Book = Sales[::2]
Price = Sales[1::2]
Book_after=[]
cost=[]
num=[]
print("共銷售%d本書"%len(Book))
print("\n")
for i in range(len(Book)):
  if Book[i] not in Book_after:
      Book_after.append(Book[i])
      cost.append(Price[i])
      num.append(1)
  else:
    for j in range(len(Book_after)):
      if Book[i]==Book_after[j]:
        cost[j]+=Price[i]
        num[j]+=1

# print(Book_after,num)

print("每本書總共銷售金額：")
for i in range(len(Book_after)):
  print(Book_after[i],cost[i])

print("\n")

"""
for i in range(len(Book_after)):
  print(Book_after[i],num[i])

print("\n")
"""

print("銷售數量最多的書：")
for i in range(len(Book_after)):
  if num[i]==max(num):
    print(Book_after[i],":",num[i],"本")
    break

print("\n")
min=len(Book_after[0])
minnum=0

print("書名最短的書：")
for i in range(len(Book_after)):
  if len(Book_after[i])<min:
    min=len(Book_after[i])
    minnum=i

print(Book_after[minnum],":",min,"字")
```
