```python
student_scores = {"Alice": 85, "Bob": 90, "Charlie": 78, "David": 92}

while True:
    action = input("輸入 '查詢' 查詢分數，'新增' 新增/更新分數，或 'exit' 結束：")

    if action == "查詢":
        name = input("請輸入學生姓名：")
        if name in student_scores:
            print(f"{name} 的分數是 {student_scores[name]}。")
        else:
            print(f"{name} 不在名單內。")

    elif action == "新增":
        name = input("請輸入學生姓名：")
        score = int(input(f"請輸入 {name} 的分數："))
        if name in student_scores:
            student_scores[name] = score
            print(f"{name} 的分數已更新為 {score}。")
        else:
            student_scores[name] = score
            print(f"{name} 已新增，分數為 {score}。")
```

```python
student_scores = {"Alice": 85, "Bob": 90, "Charlie": 78, "David": 92}

while True:
    action = input("輸入 '查詢' 查詢分數，'新增' 新增/更新分數，或 'exit' 結束：")

    if action == "查詢":
        name = input("請輸入學生姓名：")
        if name in student_scores:
            print(f"{name} 的分數是 {student_scores[name]}。")
        else:
            print(f"{name} 不在名單內。")

    elif action == "新增":
        name = input("請輸入學生姓名：")
        score = int(input(f"請輸入 {name} 的分數："))
        if name in student_scores:
            student_scores[name] = score
            print(f"{name} 的分數已更新為 {score}。")
        else:
            student_scores[name] = score
            print(f"{name} 已新增，分數為 {score}。")

    elif action == "exit":
        print("操作結束，感謝使用！")
        break

    else:
        print("無效的指令，請重新輸入。")
```
