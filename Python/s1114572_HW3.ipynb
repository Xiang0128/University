```python
stars = int(input("請輸入一個大於0的奇數："))
while(stars<=0 or stars%2==0):
  stars = int(input("請輸入一個大於0的奇數："))
for i in range(int(stars//2)+1):
  count = int(stars//2)
  for j in range(count):
    print(" ",end="")
  for j in range(2*i+1):
    print("*",end="")
  for j in range(count):
    print(" ",end="")
  print("")
```

```python
stars = int(input("請輸入一個大於0的奇數："))
while(stars<=0 or stars%2==0):
  stars = int(input("請輸入一個大於0的奇數："))
for i in range(int(stars//2)+1):
  count = int(stars//2)-i
  for j in range(count):
    print(" ",end="")
  for j in range(2*i+1):
    print("*",end="")
  for j in range(count):
    print(" ",end="")
  print("")
```
