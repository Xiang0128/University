```python
import matplotlib.pyplot as plt

years = [2020, 2021, 2022, 2023, 2024]
sales = [80, 95, 110, 130, 150]

plt.figure(figsize=(10, 6))
plt.plot(years, sales, marker='o', color='b', label='Annual Sales')

for x, y in zip(years, sales):
    plt.text(x, y + 2, f'{y}K', ha='center', fontsize=10, color='black')

average_sales = sum(sales) / len(sales)

plt.axhline(y=average_sales, color='r', linestyle='--', label=f'Average Sales: {average_sales:.1f}K')

plt.grid(True, linestyle='--', alpha=0.7)

plt.legend(loc='upper left', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

years = [2020, 2021, 2022, 2023, 2024]
sales = [80, 95, 110, 130, 150]

plt.figure(figsize=(10, 6))
plt.plot(years, sales, marker='o', color='b', label='Annual Sales')

for x, y in zip(years, sales):
    plt.text(x, y + 2, f'{y}K', ha='center', fontsize=10, color='black')

average_sales = sum(sales) / len(sales)

plt.axhline(y=average_sales, color='r', linestyle='--', label=f'Average Sales: {average_sales:.1f}K')

plt.title('Annual Sales Trend (2020-2024)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Sales (in thousand dollars)', fontsize=12)

plt.grid(True, linestyle='--', alpha=0.7)

plt.legend(loc='upper left', fontsize=12)

plt.tight_layout()
plt.show()
```
