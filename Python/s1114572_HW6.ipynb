```python
music = {"小明", "小華", "小美", "小英"}
art = {"小美", "小英", "小強", "小麗"}

print("只選了音樂課的學生: " + str(music - art))
print("只選了美術課的學生: " + str(art - music))
print("同時選了音樂和美術課的學生: " + str(music & art))

print("至少選了一門課的學生: " + str(music | art))

print("總共有" + str(len(music | art)) + "位學生選了課。")
```
